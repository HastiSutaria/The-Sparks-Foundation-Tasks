# -> General Profit Analysis

```python
# Profit Analysis of category
category_s=data.groupby('Category').Sales.sum()
category_p=data.groupby('Category').Profit.sum()
plt.style.use('seaborn')
category_s.plot(kind='bar',figsize=(10,5),legend=True)
category_p.plot(kind='bar',figsize=(10,5),color='coral',legend=True)
plt.xlabel('Category',fontsize=15)
plt.ylabel('Total Sales/Profits', fontsize=15)
plt.title('Sales vs Profits', fontsize=15)
plt.show()
```

```python
# Profit Analysis of Sub-category
category_s=data.groupby('Sub-Category').Sales.sum()
category_p=data.groupby('Sub-Category').Profit.sum()
plt.style.use('seaborn')
category_s.plot(kind='bar',figsize=(10,5),legend=True)
category_p.plot(kind='bar',figsize=(10,5),color='coral',legend=True)
plt.xlabel('Sub-Category',fontsize=15)
plt.ylabel('Sales/Profits',fontsize=15)
plt.title('Sub-Category Sales vs Profits',fontsize=15)
plt.show()
```

```python
# Profit Analysis of State
category_s=data.groupby('State').Sales.sum()
category_p=data.groupby('State').Profit.sum()
plt.style.use('seaborn')
category_s.plot(kind='bar',figsize=(13,7),legend=True)
category_p.plot(kind='bar',figsize=(13,7),color='coral',legend=True)
plt.xlabel('States',fontsize=15)
plt.ylabel('Total Sales/Profits',fontsize=15)
plt.title('States Sales vs Profits',fontsize=15)
plt.show()
```

```python
# Profit Analysis of Region
plt.figure(figsize=(7,7))
plt.title('Region')
plt.pie(data['Region'].value_counts(),labels=data['Region'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

```python
# Profit Analysis of Segment
plt.figure(figsize=(7,7))
plt.title('Segment')
plt.pie(data['Segment'].value_counts(),labels=data['Segment'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

```python
# Profit Analysis of Ship Mode
plt.figure(figsize=(7,7))
plt.title('Ship Mode')
plt.pie(data['Ship Mode'].value_counts(),labels=data['Ship Mode'].value_counts().index,autopct='%1.1f%%')
plt.show()
```
